```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
data = pd.read_csv('survey_lung_cancer.csv')
```

```python
data['GENDER'] = data['GENDER'].replace({'M': 0, 'F': 1})
```

```python
print ("Dataset Lenght:: ", len(data))
```

```python
print ("Dataset Shape:: ", data.shape)
```

```python
first_12_rows = data.head(12)
display(first_12_rows)

last_5_rows = data.tail(5)
display(last_5_rows)

num_rows, num_columns = data.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
```

```python
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Separating the Target Variable

X = data.values[:, 0:15]
Y = data.values[:, 15]
```

```python
type(X)
```

```python
data.isnull().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size = 0.3, random_state = 100)
```

```python
clf_entropy = DecisionTreeClassifier(criterion = "entropy", random_state = 100,
 max_depth=3, min_samples_leaf=5)
clf_entropy.fit(X_train, y_train)
```

```python
y_pred_en = clf_entropy.predict(X_test)
y_pred_en
```

```python
print ("Accuracy is "), accuracy_score(y_test,y_pred_en)*100
```

```python
# Heatmap
data['LUNG_CANCER'] = data['LUNG_CANCER'].map({'YES': 1, 'NO': 0})
plt.figure(figsize=(12,8))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```

```python
model = LogisticRegression(max_iter=10000)
model.fit(X_train, y_train)

predictions = model.predict(X_test)

first_10_actual = y_test[:10]
first_10_predicted = predictions[:10]

print("First 10 actual values:", first_10_actual)
print("First 10 predicted values:", first_10_predicted)
```

```python
#Confusion Matrix
cm = confusion_matrix(y_test, predictions)
print("Confusion Matrix:")
print(cm)

#  Precision, recall, and F1 score
precision = precision_score(y_test, predictions, pos_label='YES')
recall = recall_score(y_test, predictions, pos_label='YES')
f1 = f1_score(y_test, predictions, pos_label='YES')

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```
